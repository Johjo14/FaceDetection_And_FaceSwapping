```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
#Chargement des classificateurs en cascade pré-entrainés
cascade_face = cv.CascadeClassifier("haarcascade_frontalface_default.xml")
cascade_eyes = cv.CascadeClassifier("haarcascade_eye.xml")
#On créer deux variables correspondant aux photos
portrait_img = "ImagesData/portrait_1.jpg"
couple_img = "ImagesData/couple_macron.jpg"
```

```python
#On affiche la photo de la personne
img = cv.imread(portrait_img)
if img is None:
    #Si image n'est pas trouver alors j'affiche un message d'erreur
    print("Erreur: l'image n'a pas été charger...")
else:
    #Sinon j'affiche l'image
    plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.title("Image portrait")
    plt.axis("off")
    plt.show()
    
```

```python
#J'affiche la photo du couple Macron
img = cv.imread(couple_img)
if img is None:
    #Si image n'est pas trouver alors j'affiche un message d'erreur
    print("Erreur: l'image n'a pas été charger...")
else:
    #Sinon j'affiche l'image
    plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.title("Image chargée du couple Macron")
    plt.axis("off")
    plt.show()
    
```

```python
#On Charge les images
#Le portrait en photo
img = cv.imread(portrait_img)
#On converti en niveau gris pour plus de facilité pour le fichier cascade
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

#Execution de la detection du visage
#detectMultiScale(image, scale factor, number of neighbors(nombre de voisins)
faces = cascade_face.detectMultiScale(gray, 1.1, 3) 
eyes = cascade_eyes.detectMultiScale(gray, 1.1, 3)


#J'affichage des coordonnées[liste]de la position du visage
for face in faces:
    print(face)

print("\n")

#J'affiche le ou les visages
for face in faces:
    x, y, w, h = face
    cv.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 1)

#J'affiche également les yeux
for (ex, ey, ew, eh) in eyes:
    #Je dessine le rectangle autour des yeux sur l'image principale
    cv.rectangle(img, (ex, ey), (ex+ew, ey+eh), (255, 0, 0), 1) 
#On affiche l'image avec les rectangles qui encadrent le ou les visage
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.show()
```

```python
#On Charge les images
#Photo de portrait
img = cv.imread(portrait_img)
if img is None:
    print("Erreur: l'image n'a pas été correctement chargée...")
else:
    #Je converti en niveau de gris pour plus de facilité pour le fichier cascade
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    #J'éxécute la detection du visage
    faces = cascade_face.detectMultiScale(gray, 1.1,3)
 
    #J'affiche les coordonnées de la position du visage
    for face in faces:
        print(face)
    print("\n")
    #Je dessine les rectangles
    face_imgs = []
    for face in faces:
        x, y, w, h = face
        cv.rectangle(img, (x,y), (x+w, y+h), (0, 255, 0), 1)
        #Je fais une extraction du visage de l'image principale avec OpenCV et Numpy(y vaut row, x vaut col)
        face_img = img[y:y + h, x:x + w]
        #J'affiche le visage0, visage1, visage2...
        face_imgs.append(face_img)
        
    #J'affiche l'image principale avec les rectangles
    plt.figure(figsize = (10, 10))
    plt.subplot(1, len(face_imgs) + 1, 1)
    plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.title(f"Image portrait")
    plt.axis("off")

    #J'affiche les visages 
    for i, face_img in enumerate(face_imgs):
        plt.subplot(1, len(face_imgs)+ 1, i + 2)
        plt.imshow(cv.cvtColor(face_img, cv.COLOR_BGR2RGB))
        plt.title(f"Image portrait")
        plt.axis("off")
    plt.show()
```

```python
#On Charge les images
#Photo du couple Macron
img = cv.imread(couple_img)
#je converti en niveau gris pour plus de facilité pour le fichier cascade
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

#j'éxecute la detection du visage
#detectMultiScale(image, scale factor, number of neighbors(nombre de voisins)
faces = cascade_face.detectMultiScale(gray, 1.1, 8)
eyes = cascade_eyes.detectMultiScale(gray, 1.1, 5)

#J'affichage des coordonnées[liste]de la position du visage
for face in faces:
    print(face)

print("\n")

#Je déssine des rectangles
for face in faces:
    x, y, w, h = face
    cv.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
#J'affiche les yeux
for (ex, ey, ew, eh) in eyes:
    #Je dessine le rectangle autour des yeux sur l'image principale
    cv.rectangle(img, (ex, ey), (ex+ew, ey+eh), (255, 0, 0), 2) 
    
#J'affiche l'image avec les rectangles qui encadrent le ou les visage
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.show()
```

```python
#On Charge les images
#Photo du couple Macron
img = cv.imread(couple_img)
if img is None:
    print("Erreur: l'image n'a pas été correctement chargée...")
else:
    #Je converti en niveau gris pour plus de facilité pour le fichier cascade
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    #J'éxécute la detection du visage
    faces = cascade_face.detectMultiScale(gray, 1.1,8) #parametre qui va jouer sur la detection des visage
    #J'affiche les coordonnées de la position du visage
    for face in faces:
        print(face)
    print("\n")
    #Je dessine les rectangles
    face_imgs = []
    for face in faces:
        x, y, w, h = face
        cv.rectangle(img, (x,y), (x+w, y+h), (0, 255, 0), 2)
        #Je fais une extraction du visage de l'image principale avec OpenCV et Numpy(y vaut row, x vaut col)
        face_img = img[y:y + h, x:x + w]
        #J'affiche le visage0, visage1, visage2...
        face_imgs.append(face_img)
        
    #J'affiche l'image principale avec les rectangles
    plt.figure(figsize = (10, 10))
    plt.subplot(1, len(face_imgs) + 1, 1)
    plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.title(f"Image du couple Macron")
    plt.axis("off")

    #J'affiche les visages extraits
    for i, face_img in enumerate(face_imgs):
        plt.subplot(1, len(face_imgs)+ 1, i + 2)
        plt.imshow(cv.cvtColor(face_img, cv.COLOR_BGR2RGB))
        plt.title(f"Fragment de l'image")
        plt.axis("off")
    plt.show()
    
```

<h2>Extraction des visages d'une photo</h2>

```python
#Importation des modules
import cv2 as cv
import sys
```

```python
#Je charge le classificateur en cascade pré-entrainé
face_cascade = cv.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
#Je charge l'image
img =cv.imread("ImagesData/couple_acteurs.jpg")
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python

#J'execute la détéction de visage
faces = face_cascade.detectMultiScale(img_gray, 1.1, 8)

#Je vérifie le nombre de visages
if len(faces) != 2:
    sys.exit("La photo doit avoir 2 visages pour que ca marche...")

#Je récupèrere les dimensions de chaque visage
x1, y1, w1, h1 = faces[0]
x2, y2, w2, h2 = faces[1]
#Je fais une extaction des deux visages de l'image
face1 = img[y1:y1 + h1, x1:x1 + w1]
face2 = img[y2:y2 + h2, x2:x2 + w2]

#Je redimensionne face2 aux dimensions de face1 et vice versa
face2 = cv.resize(face2, (w1, h1))
face1 = cv.resize(face1, (w2, h2))

#Je remplace face2 par face1
img[y1:y1 + h1, x1:x1 + w1] = face2
#Je remplace face1 par face2
img[y2:y2 + h2, x2:x2 + w2] = face1

#J'affiche les regions composant les visage
cv.imshow("échange", img)
cv.waitKey(0)
cv.destroyAllWindows()

# #On affiche des visages
# cv.imshow("face1", face1)
# cv.imshow("face2", face2)
# cv.waitKey(0)
# cv.destroyAllWindow(0)
```
